```python
from collections import defaultdict
```

```python
from treepath import TreePath
```

```python
def rot90(pos):
    
    return {k:(-y,x) for k,(x,y) in pos.items()}

rot90({'a' : (1,2)})
        
```

```python
wdir = os.path.abspath(".")
file_tree_paths=[]
for root, dirs, files in os.walk(wdir):
    #print (root, dirs, files)
    for d in dirs:
        file_tree_paths.append((root.split(os.sep),d))
    for f in files:
        file_tree_paths.append((root.split(os.sep),f))
```

```python
T = TreePath({})
for p in file_tree_paths:
    T.add(*(p))
```

```python
#T.data = T.data['']
```

```python
import matplotlib.pyplot as plt
import networkx as nx
import pygraphviz as pgv
```

```python
#g= nx.DiGraph()


    
labels = dict({n:d.get('data',d).get('label') for n,d in g.nodes(data=True)})
#inv_labels = {v:k for k,v in labels.items()}

fig = plt.figure(figsize=(15,40))
pos = nx.nx_agraph.pygraphviz_layout(g, prog="dot")
root_node = [n for n,d in g.in_degree() if d==0][0]
sub_nodes = nx.descendants(g, root_node)
#sub_nodes = nx.ancestors(g, inv_labels['home'])
sub_nodes.add(root_node)
sub_labels = {k:v for k,v in labels.items() if k in sub_nodes}
subgraph = g.subgraph(sub_nodes)
spos = nx.nx_agraph.pygraphviz_layout(subgraph, prog="dot")
nx.draw_networkx(subgraph, pos=rot90(spos), labels=sub_labels)
```
